```python
#HEARTHRATE

df_heartrate['date'] = pd.to_datetime(df_heartrate['date'])
df_heartrate['time'] = pd.to_datetime(df_heartrate['time'])
#put each unique day into a list
unique_days_list = df_heartrate['date'].dt.date.unique().tolist()
#plot each day


def plot_data_per_day(dataframe, date_column, data_column, y):
    """
    Creates separate line plots for each day based on the specified date and data columns.

    Parameters:
    - dataframe: pandas DataFrame
    - date_column: str, the name of the column containing dates
    - data_column: str, the name of the column containing data points
    """


    # Group by date and plot data points for each day
    for day, group in dataframe.groupby(dataframe[date_column].dt.date):
        plt.figure()
        plt.plot(group[data_column], group[y], marker='o')
        plt.title(f'Plot for {day}')
        plt.xlabel('Date')
        plt.ylabel('Data Points')
        plt.show()

# Example usage:
# Assuming you have a DataFrame named 'your_data' with a column 'days'
# containing dates and a column 'datapoints' containing data points,
# and you want to create separate line plots for each day
plot_data_per_day(df_heartrate, 'date', 'time','heartRate')
```
